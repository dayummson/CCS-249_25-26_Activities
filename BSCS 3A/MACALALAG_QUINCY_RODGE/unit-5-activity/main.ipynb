```python

CORPUS = [
    "Climate change affects global weather patterns significantly.",
    "Rising sea levels threaten coastal cities worldwide.",
    "Deforestation contributes to higher carbon dioxide levels.",
    "Renewable energy sources like solar and wind are essential.",
    "Solar panels convert sunlight into clean electricity.",
    "Wind turbines generate power from moving air currents.",
    "Greenhouse gases trap heat in Earth's atmosphere.",
    "Carbon emissions from factories pollute the air quality.",
    "Electric vehicles reduce dependence on fossil fuels.",
    "Fossil fuels such as coal and oil are finite resources.",
    "Biodiversity loss impacts ecosystems and food chains.",
    "Conservation efforts protect endangered species habitats.",
    "Recycling reduces waste and conserves natural resources.",
    "Sustainable agriculture promotes soil health and water conservation.",
    "International agreements aim to limit global warming."
]
```

```python
tokenized_corpus = [re.findall(r'\b\w+\b', sentence.lower()) for sentence in CORPUS]
vocab = list(set([word for sentence in tokenized_corpus for word in sentence]))

word_to_id = {word: i for i, word in enumerate(vocab)}
id_to_word = {i: word for i, word in enumerate(vocab)}
```

2. (10 points) Which vector in the word-context matrix is the longest/highest
magnitude? What does it mean?


- The vector with the highest magnitude corresponds to the word that has the most frequent and diverse co-occurrences within the ±3 word window across the entire corpus. In natural language processing, this is almost always a highly frequent grammatical "stop word" (like and, to, or are) or a repeating thematic keyword (like fuels or resources). A high magnitude means the word acts as a central hub, appearing often and sharing close proximity with many other words in the text.

```python
WINDOW_SIZE = 3
VOCAB_SIZE = len(vocab)
matrix = np.zeros((VOCAB_SIZE, VOCAB_SIZE))

for sentence in tokenized_corpus:
    for i, target_word in enumerate(sentence):
        target_id = word_to_id[target_word]
        start = max(0, i - WINDOW_SIZE)
        end = min(len(sentence), i + WINDOW_SIZE + 1)

        for j in range(start, end):
            if i != j:
                context_id = word_to_id[sentence[j]]
                matrix[target_id][context_id] += 1
```

3. Calculate the cosine similarity of the following vectors, and describe how
each word is similar linguistically:

a. (climate, global): Linguistically, both function frequently as adjectives modifying broad environmental concepts (e.g., "climate change", "global warming", "global weather"). They share semantic space in discussions about planetary-scale phenomena, meaning they will co-occur with similar context words like change, warming, or verbs related to impact.

b. (solar, wind): These are highly similar both syntactically and semantically. They act as noun adjuncts or adjectives describing renewable energy mechanisms (e.g., "solar panels", "wind turbines"). Because they are often mentioned together (as seen in Sentence 4: "solar and wind"), they share almost identical contextual neighborhoods in an environmental text.

c. (fossil, coal): Semantically, both relate to non-renewable carbon-based energy. Linguistically, "fossil" is usually a modifier ("fossil fuels"), while "coal" is a specific instance or noun. They share a high contextual overlap because they appear in identical discussions regarding emissions, fuels, and finite resources.

```python
def cosine_similarity(vec1, vec2):
    if norm(vec1) == 0 or norm(vec2) == 0:
        return 0.0
    return np.dot(vec1, vec2) / (norm(vec1) * norm(vec2))

pairs = [('climate', 'global'), ('solar', 'wind'), ('fossil', 'coal')]

print("3. Cosine Similarities:")
for w1, w2 in pairs:
    v1 = matrix[word_to_id.get(w1, 0)]
    v2 = matrix[word_to_id.get(w2, 0)]
    sim = cosine_similarity(v1, v2)
    print(f"Similarity ({w1}, {w2}): {sim:.4f}")
```

```python
magnitudes = norm(matrix, axis=1)
longest_id = np.argmax(magnitudes)
longest_word = id_to_word[longest_id]
longest_vector = matrix[longest_id]

print(f"The longest vector is for the word: '{longest_word}'\n")

print("Full vector array (showing counts for all words in vocab):")
print(longest_vector)

print("\nNon-zero context words and their co-occurrence counts:")
for i, count in enumerate(longest_vector):
    if count > 0:
        print(f" - {id_to_word[i]}: {int(count)}")
```
