# Exercise for Unit 4.1 Naïve Bayes

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Util to get non empty string
from utils import utils

# ----------------------------
# Training Dataset
# ----------------------------
from dataset import dataset

# ----------------------------
# Convert text to Bag of Words
# ----------------------------
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform([item["text"] for item in dataset])

# ----------------------------
# Extract labels
# ----------------------------
labels = [item["label"] for item in dataset]

# ----------------------------
# Train Multinomial Naïve Bayes
# ----------------------------
model = MultinomialNB()
model.fit(X_train, labels)

# ----------------------------
# Test Sentences
# ----------------------------

user_sentence = utils.get_str("Enter a test sentence: ")
test_sentences = [
    "Limited offer, click here!",
    "Meeting at 2 PM with the manager.",

    user_sentence
]

X_test = vectorizer.transform(test_sentences)
predictions = model.predict(X_test)

# ----------------------------
# Output Results
# ----------------------------
for sentence, prediction in zip(test_sentences, predictions):

    print(f"Sentence: {sentence}")
    print(f"Predicted Class: {prediction}")
    print()
```
